```python
import numpy as np 
from matplotlib import pyplot as plt 

x = np.arange(1,5) 
y = 2 * x  
plt.title("Matplotlib demo") 
plt.xlabel("x axis") 
plt.ylabel("y axis") 
plt.plot(x,y) 
plt.show()
```

```python
from matplotlib import pyplot as plt 
import numpy as np  
   
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a, bins = [0,20,40,60,80,100]) 
plt.title("histogram") 
plt.show()
```
